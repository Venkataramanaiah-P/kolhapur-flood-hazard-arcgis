```python
from arcpy.sa import *
import arcpy
import os

# Create output folder if not exists
os.makedirs(r"D:\GIS_Projects\Kolhapur_P8", exist_ok=True)

arcpy.env.overwriteOutput = True

dem = Raster("dem_clasifie")
rain = Raster("Rainfall")
pop = Raster("Populatioj")
slope = Raster("slope_cla")
ndvi = Raster("ndvi_classi")

hazard = (dem * 0.30) + (rain * 0.25) + (pop * 0.20) + (slope * 0.15) + (ndvi * 0.10)

# Save to project geodatabase instead
hazard.save("Flood_Hazard_P8_final")

print("Done!")
```
